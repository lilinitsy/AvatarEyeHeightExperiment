```python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt

positions = pd.read_csv("C:\\Users\\vcant\\Desktop\\AvatarEyeHeightExperiment4.24\\AvatarEyeHeightExperiment\\headset_position.txt", header = None)
rotations = pd.read_csv("C:\\Users\\vcant\\Desktop\\AvatarEyeHeightExperiment4.24\\AvatarEyeHeightExperiment\\headset_rotation.txt", header = None)

print("Num datapoints rotation: ", rotations.shape)
print("Num datapoints position: ", positions.shape)

print("rotations[2]", rotations.iloc[2, 0])

fig1 = plt.figure(figsize = (16, 16))
ax0 = fig1.add_subplot(1, 1, 1)

y_rotations = rotations.iloc[:, 0]
print("Num datapoints: ", len(y_rotations))

xgrid_rot = np.linspace(0, len(y_rotations), len(y_rotations))

ax0.scatter(xgrid_rot, y_rotations, label = "y rotation of camera")

y_positions = positions.iloc[:, 1] # forward and backwards is y here
xgrid_pos = np.linspace(0, len(y_positions), len(y_positions))
ax0.scatter(xgrid_pos, y_positions, label = "y position of camera relative to origin")

ax0.legend()
```
